```python
df2["Athlete Age"] = df2["Athlete Age"].astype(int)
```

```python
df2["Athlete Age"] = df2["Athlete average speed"].astype(float)
```

```python
df2 = df2.rename(columns={"year":"Year",
                          'Event dates': "Race_day",
                          'Event name': "Race_name",
                          "Event distance/length": "Race_length",
                          "Event number of finishers": "Race_number_of_finishers",
                          "Athlete performance": "Athlete_performance",
                          "Athlete gender":"Athlete_gender",
                          "Athlete average speed": "Athelete_average_speed",
                          "Athlete ID": "Athelete_id"
                          })

df2
```

```python
df3 = df2[["Race_day","Race_name","Race_length","Race_number_of_finishers","Athelete_id","Athlete_gender", "Athlete Age","Athlete_performance","Athelete_average_speed"]]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.lmplot(
    data=df3,
    x="Athlete Age",
    y="Athelete_average_speed",
    hue="Athlete_gender",
    height=6,
    aspect=1.5,
    scatter_kws={"alpha": 0.6},
    line_kws={"linewidth": 2}
)

plt.title("Relationship Between Age and Average Speed by Gender")
plt.xlabel("Athlete Age")
plt.ylabel("Average Speed")
plt.tight_layout()
plt.show()
```
